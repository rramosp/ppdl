- **task 1**: create a function that return the likelihood of a beta distribution wrt some data. this function must have two parameters (a,b for the beta distribution)

$$\mathcal{L}(a,b) = P(d|a,b)$$

- **task 2**: create a function that returns the double integral over a,b with a fixed data set --> the likelihood is not a probability

$$Z = \int_a \int_b P(d|a,b)\text{d}a\text{d}b$$

- **task 3**: create a function that optimizes the likelihood

$$\underset{a,b}{\text{argmax}}\;\;\mathcal{L}(a,b)$$

- **task 4**: create a function that given priors on the parameters $P(a)$ and $P(b)$ it computes

$$P(d) = \int_a \int_b P(d|a,b)P(a)P(b)\text{d}a\text{d}b$$

making emphasis on that this is $P(d)$ ... we are obtaining the marginal probability of our specific dataset by using the conditional and the marginals of the parameters (which as have seen in bayes videos, are also callsed "the priors").
