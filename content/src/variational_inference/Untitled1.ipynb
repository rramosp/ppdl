```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import quad
%matplotlib inline
```

```python
f = lambda x,a=1: np.cos(a+x)/(np.sin(a+1)-np.sin(a))

```

```python
x = np.linspace(0,1,100)
```

```python
a_range = np.linspace(-np.pi/2,np.pi/2-1,10)
for a_val in a_range:
    fx = f(x, a=a_val)
    integral_check = quad(lambda x: f(x, a=a_val), 0,1)[0]
    plt.plot(x, fx, label=f"{a_val:.3f}    min fx {np.min(fx):.3f}   integral check {integral_check:.3f}")

plt.grid();
plt.legend(loc='center left', bbox_to_anchor=(1, 0.5))
```
