```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
```

```python
w = np.linspace(0,1,100)

l_mean = 2+(w)**2
l_std  = .1+.5*w
```

```python
plt.plot(w, l_mean)
plt.plot(w, l_mean + l_std, color="black")
plt.plot(w, l_mean - l_std, color="black")
```

```python
p_l_given_w = lambda w: stats.norm(loc=2+w**2, scale=.1+w/2).pdf

lr = np.linspace(1,4,100)

plt.plot(lr, p_l_given_w(w=.3)(lr))
plt.plot(lr, p_l_given_w(w=.8)(lr))
```

```python
from scipy.integrate import quad, dblquad

p_w = lambda w: 1.
p_conditional = lambda l,w: stats.norm(loc=2+w**2, scale=.1+w/2).pdf(l)
p_lw = np.vectorize(lambda l,w: p_l_given_w(l,w)*p_w(w))

dblquad(p_lw, 0, 1, lambda w: 0, lambda l: 5)
```

```python
p_w = lambda w: stats.norm(loc=.5,scale=.2).pdf(w)
p_w = lambda w: stats.beta(3,2).pdf(w)
p_conditional = np.vectorize(lambda l,w: stats.norm(loc=2+w**2, scale=.1+w/2).pdf(l))
p_lw = np.vectorize(lambda l,w: p_conditional(l,w)*p_w(w))
```
