```python
from scipy import stats

xr = np.linspace(50,200, 100)
fun = lambda x: stats.beta(a=15, b=5).pdf( (x-50)/150)
plt.plot(xr, fun(xr))
```

```python
from scipy import stats

xr = np.linspace(0,5, 100)
fun = lambda x: stats.beta(a=2, b=4).pdf( (x)/5)
plt.plot(xr, fun(xr))
```
